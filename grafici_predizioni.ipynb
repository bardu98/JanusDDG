```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, spearmanr
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Results/Result_m28_test_set.csv')

# Creiamo il grafico scatter
plt.figure(figsize=(8, 6))
ax = sns.scatterplot(data=df, x='DDG', y='DDG_JanusDDG', palette='coolwarm')

# Aggiungiamo la bisettrice y = x
x_vals = np.linspace(df['DDG'].min(), df['DDG'].max(), 100)
plt.plot(x_vals, x_vals, linestyle='dashed', color='red')

# Miglioriamo il layout
plt.xlabel("DDG")
plt.ylabel("DDG_JanusDDG")
plt.legend(title="Distance in Seq", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
